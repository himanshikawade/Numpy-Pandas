```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(100)
y=np.random.rand(100)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c="purple",marker="x")
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("X VS Y")
plt.grid()
```

```python
plt.figure(figsize=(3,2))
plt.plot(x,y,c="red")
plt.xlabel("X data")
plt.ylabel("Y data")
plt.title("this is Sin wave graph")
plt.grid()
```
